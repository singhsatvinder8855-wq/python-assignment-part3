```python
# Task 2: API Integration:
import requests

print("Requests is working.")
# Step 1: Fetch and Display Products:
# Fetching the data from dummyjson

url = "http://dummyjson.com/products?limit=20"

response = requests.get(url)
products_data = response.json()

print("Products Data fetched successfully")

# Extracting Product list

products = products_data["products"]

print("Total Products are: ", len(products))

# Creating and Printing the table:

print("\nID  | Title                                     | Category      | Price    | Rating")
print("----|-------------------------------------------|---------------|----------|----------")

for p in products:
    print(f'{p["id"]:<3} | {p["title"]:<41} | {p["category"]:<13} | $ {p["price"]:<8} | {p["rating"]}')
```

```python
# STep 2: Filter and Sort products
# Filtering and sorting the products above 4.5

high_rated = []

for p in products:
    if p["rating"] >= 4.5:
        high_rated.append(p)
        
print("\nProducts above 4.5 are: ")

# Sort the products by price (descending order)

for i in range(len(high_rated)):
    for j in range(i + 1, len(high_rated)):
        if high_rated[i]["price"] < high_rated[j]["price"]:
            temp = high_rated[i]
            high_rated[i] = high_rated[j]
            high_rated[j] = temp

# Print the sorted products
print("\nSorted products by price (descending):")
for p in high_rated:
    print(f"ID: {p['id']}, Title: {p['title']}, Price: {p['price']}, Rating: {p['rating']}")
```

```python
# Step 3: Search by category:

url2 = "https://dummyjson.com/products/category/laptops"

response2 = requests.get(url2)
data2 = response2.json()

print("\nLaptops found: ")

for p in data2["products"]:
    print(f'{p["title"]} - ${p["price"]}')
    
# Step 4: Post Request (Send new product)

url3 = "https://dummyjson.com/products/add"

new_product = {
  "title": "My Custom Product",
  "price": 999,
  "category": "electronics",
  "description": "A product I created via API"
}

response3 = requests.post(url3, json=new_product)

print("\nPost Response: \n")

data = response3.json()

print("ID:", data["id"])
print("Title:", data["title"])
print("Price:", data["price"])
print("Category:", data["category"])
print("Description:", data["description"])
```

```python
# Task 3: Exception Handling
# Part A: Safe Divide function

def safe_divide(a, b):
    try:
        return a/b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
    except TypeError:
        return "Error: Invalid input types.\n"
    
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))

# Part B: File Reader

def read_file_safe(filename):
    try:
        file = open(filename, "r", encoding="utf-8")
        content = file.read()
        file.close()
        return content
    except FileNotFoundError:
        print(f"Error: No file named '{filename}' found.")
    finally:
        print("File operation attempt complete.")
        
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))
```

```python
# Part D: Product lookup with input validation

import requests

while True:
    user_input = input("Enter Product ID (1-100) or 'quit' to exit: ")
    
    if user_input.lower() == "quit":
        print("Exiting...")
        break
    
    if user_input.isdigit():
        product_id = int(user_input)
        
        if 1 <= product_id <= 100:
            try:
                url = f"https://dummyjson.com/products/{product_id}"
                response = requests.get(url, timeout=10)
                
                if response.status_code == 200:
                    product = response.json()
                    print("Product Name: ", product["title"])
                    print("Price: ", product["price"])
                    
                elif response.status_code == 404:
                    print("Product not found")
                    
                else:
                    print(f"Unexpected status code: {response.status_code}")
                    
            except requests.exceptions.ConnectionError:
                print("Connection failed. Please check your internet connection.")
    
            except requests.exceptions.Timeout:
                print("Request time out. Try again later.")
    
            except Exception as e:
                print("Error: ", e)
                
        else:
            print("Please enter a number between 1 to 100.")
            
    else:
        print("Invalid input. Please enter a number.")
```

```python
# Task 4: Logging errors to file

from datetime import datetime
import requests

# Step 1: Function to write log
def write_log(function_name, error_type, message):
    with open("error_log.txt", "a", encoding="utf-8") as file:
        time_now = datetime.now()
        log_line = f"[{time_now}] Error in {function_name}: {error_type} - {message}\n"
        file.write(log_line)
        
# Step 2: Trigger the Connection Error by Fake Url
try:
    bad_url = "https://the-host-is-busy-abc.com/api"
    requests.get(bad_url, timeout=5)
    
except requests.exceptions.ConnectionError as e:
    print("Connection Error Occurred.")
    write_log("fetch_products", "ConnectionError", str(e))
    
# Step 3: Trigger http error
try:
    url = "https://dummyjson.com/products/999"
    response = requests.get(url, timeout=5)
    
    if response.status_code != 200:
        print("HTTP error occured")
        write_log("lookup_product", "HttpError", "Product Id 999 not found")
        
except Exception as e:
    write_log("lookup_product", "Exception", str(e))
    
# Step 4: Read and Print the log file
print("\nError logs from file:\n")

with open("error_log.txt", "r", encoding="utf-8") as file:
    logs = file.read()
    print(logs)
```
